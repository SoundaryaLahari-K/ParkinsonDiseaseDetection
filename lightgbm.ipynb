```python
import lightgbm as lgb
import pandas as pd
from sklearn.discriminant_analysis import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python

# # load data
# data = pd.read_csv('pd_speech_features.csv')
# X = data.drop('class', axis=1)
# y = data['class']
data2=pd.read_csv('parkinsons.csv')
X = data2.drop(['name','status'], axis=1)
y = data2['status']
# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# create a LightGBM dataset from the training data
train_data = lgb.Dataset(X_train_scaled, label=y_train)
```

```python

# set hyperparameters for the LightGBM model
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'goss',
    'num_leaves': 31,
    'learning_rate': 0.2,
    'num_iterations':1200,
    'feature_fraction': 0.9,
    'max_depth': 2,
    'max_bins':150,
}

```

```python

# train the LightGBM model using cascading
num_round = 15
bst = lgb.train(params, train_data, num_round)

# make predictions on the test set
y_pred = bst.predict(X_test)
y_pred = [round(x) for x in y_pred]
```

```python
# calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from lightgbm import LGBMClassifier
from sklearn.ensemble import StackingClassifier
from xgboost import XGBClassifier

base_model=[('lgb',LGBMClassifier()),('lgb2', LGBMClassifier( 
        objective= 'binary',
        metric= 'binary_logloss',boosting_type= 'goss',
    num_leaves= 31,
    learning_rate= 0.2,
    num_iterations=1200,
    feature_fraction= 0.9,
    max_depth= 2,
    max_bins=150),X_train_scaled),('xgb',XGBClassifier())]
stacking_model = StackingClassifier(estimators=base_model, cv=5)
stacking_model.fit(X_train_scaled, y_train)
stacking_pred = stacking_model.predict(X_test)
stacking_accuracy = accuracy_score(y_test, stacking_pred)
print("Stacking Classifier Accuracy:", stacking_accuracy)
```
